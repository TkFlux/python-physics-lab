```python
import math

print("=== TK Physics Helper Toolkit ===")
print("1) Velocity (v = dx/dt)")
print("2) Acceleration (a = dv/dt)")
print("3) Kinetic Energy (KE = 1/2 m v^2)")
print("4) Gravitational Force (F = G m1 m2 / r^2)")
print("5) Photon Energy from wavelength (E = hc/λ)")

G = 6.67430e-11
h = 6.62607015e-34
c = 2.99792458e8

choice = input("Choose 1-5: ")

if choice == "1":
    dx = float(input("Enter dx (meters): "))
    dt = float(input("Enter dt (seconds): "))
    print("Velocity v =", dx/dt, "m/s")

elif choice == "2":
    dv = float(input("Enter dv (m/s): "))
    dt = float(input("Enter dt (seconds): "))
    print("Acceleration a =", dv/dt, "m/s^2")

elif choice == "3":
    m = float(input("Enter mass m (kg): "))
    v = float(input("Enter velocity v (m/s): "))
    print("Kinetic Energy KE =", 0.5*m*v**2, "J")

elif choice == "4":
    m1 = float(input("Enter mass m1 (kg): "))
    m2 = float(input("Enter mass m2 (kg): "))
    r = float(input("Enter distance r (meters): "))
    print("Gravitational Force F =", G*m1*m2/(r**2), "N")

elif choice == "5":
    lam = float(input("Enter wavelength λ (meters): "))
    E = h*c/lam
    print("Photon Energy E =", E, "J")
    print("Photon Energy E =", E/1.602176634e-19, "eV")

else:
    print("Invalid choice")
```
