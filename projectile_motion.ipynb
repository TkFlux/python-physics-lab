```python
import numpy as np
import matplotlib.pyplot as plt

# --- Inputs (change these and rerun) ---
v0 = 25.0        # initial speed (m/s)
theta_deg = 45.0 # launch angle (degrees)
y0 = 0.0         # initial height (m)
g = 9.81         # gravity (m/s^2)

theta = np.deg2rad(theta_deg)

# Time of flight from y(t)= y0 + v0*sin(theta)*t - 0.5*g*t^2 = 0
# Solve quadratic: -0.5*g*t^2 + v0*sin(theta)*t + y0 = 0
a = -0.5 * g
b = v0 * np.sin(theta)
c = y0

disc = b**2 - 4*a*c
t_flight = (-b - np.sqrt(disc)) / (2*a)  # positive root

t = np.linspace(0, t_flight, 400)

x = v0 * np.cos(theta) * t
y = y0 + v0 * np.sin(theta) * t + a * t**2  # a is negative

# Key results
range_m = x[-1]
max_height = y.max()
t_peak = t[np.argmax(y)]

print("Flight time (s):", t_flight)
print("Range (m):", range_m)
print("Max height (m):", max_height)
print("Time to peak (s):", t_peak)

# Plot trajectory
plt.figure()
plt.plot(x, y)
plt.title("Projectile Motion (No Air Resistance)")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.grid(True)
plt.show()

# Plot y(t)
plt.figure()
plt.plot(t, y)
plt.title("Height vs Time")
plt.xlabel("t (s)")
plt.ylabel("y (m)")
plt.grid(True)
plt.show()

# Plot x(t)
plt.figure()
plt.plot(t, x)
plt.title("Horizontal Position vs Time")
plt.xlabel("t (s)")
plt.ylabel("x (m)")
plt.grid(True)
plt.show()
```
