```python
def c_to_f(c):
    return (c * 9/5) + 32

def f_to_c(f):
    return (f - 32) * 5/9

def meters_to_feet(m):
    return m * 3.28084

def feet_to_meters(ft):
    return ft / 3.28084

def joules_to_ev(j):
    return j / 1.602176634e-19

def ev_to_joules(ev):
    return ev * 1.602176634e-19


print("=== TK Unit Converter ===")
print("1) Celsius → Fahrenheit")
print("2) Fahrenheit → Celsius")
print("3) Meters → Feet")
print("4) Feet → Meters")
print("5) Joules → eV")
print("6) eV → Joules")

choice = input("Choose 1-6: ")

if choice == "1":
    c = float(input("Enter °C: "))
    print("°F =", c_to_f(c))

elif choice == "2":
    f = float(input("Enter °F: "))
    print("°C =", f_to_c(f))

elif choice == "3":
    m = float(input("Enter meters: "))
    print("feet =", meters_to_feet(m))

elif choice == "4":
    ft = float(input("Enter feet: "))
    print("meters =", feet_to_meters(ft))

elif choice == "5":
    j = float(input("Enter Joules: "))
    print("eV =", joules_to_ev(j))

elif choice == "6":
    ev = float(input("Enter eV: "))
    print("Joules =", ev_to_joules(ev))

else:
    print("Invalid choice")
```
